# Standard deviation and variance

Variance $\sigma^2$ is simply **the average of the squared differences from the mean**
Standar deviation $\sigma$ is just the square root of the variance

```python
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100,20,10000)
plt.hist(incomes,50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```
